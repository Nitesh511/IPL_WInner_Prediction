```python
# Grouping the 'delivery' DataFrame by 'match_id' and 'inning',
# then summing up the 'total_runs' for each group.
total_score_df=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
# Merging the 'match' DataFrame with the 'total_score_df' DataFrame
# using 'match_id' from 'total_score_df' and 'id' from 'match' to combine relevant columns
match_df=match.merge(total_score_df[['match_id', 'total_runs']], left_on='id', right_on='match_id')
```

```python
# Defining a list of teams to work with
teams=['Sunrisers Hyderabad' ,'Kolkata Knight Riders', 'Kings XI Punjab' ,'Rajasthan Royals', 'Delhi caphtals','Chennai Super Kings',
'Mumbai Indians','Royal Challengers Bangalore']
```

```python
teams
```

```python
# Replacing 'Delhi Daredevils' with 'Delhi Capitals' in both 'team1' and 'team2' columns
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

# Replacing 'Deccan Chargers' with 'Sunrisers Hyderabad' in both 'team1' and 'team2' columns
match_df['team1']=match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df.head()
```

```python
# Filtering the 'match_df' DataFrame to only include rows where 'team1' is in the list of specified teams
match_df = match_df[match_df['team1'].isin(teams)]

# Filtering the 'match_df' DataFrame to only include rows where 'team2' is in the list of specified teams
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.columns
```

```python
# Selecting specific columns from the 'match_df' DataFrame: 'match_id', 'city', 'winner', and 'total_runs'
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df
```

```python
# Merging the 'match_df' DataFrame with the 'delivery' DataFrame on the 'match_id' column
delivery_df = match_df.merge(delivery, on='match_id')
```

```python
delivery_df.head()
```

```python
# Filtering the 'delivery_df' DataFrame to only include rows where 'inning' is equal to 2
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.head()
```

```python
delivery_df.columns
```

```python
# Calculating the cumulative sum of 'total_runs_y' for each 'match_id' to create a new column 'current_score'
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df.head(20)
```

```python
# Calculating the remaining runs ('runs_left') by subtracting 'current_score' from 'total_runs_x'
delivery_df['runs_left']=delivery_df['total_runs_x']- delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
# Calculating the number of balls remaining ('balls_left') by first determining the total balls bowled so far
# (which is calculated by multiplying the number of overs ('over') by 6 and adding the balls bowled in the current over ('ball'))
# Then, subtracting this total from 126, which is the total number of balls available in a 21-over match (6 overs per inning).
# The result is the number of balls left for the team to face in their innings.
delivery_df['balls_left']=126-(delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
# Replacing any missing (NaN) values in the 'player_dismissed' column with the string "0"
delivery_df['player_dismissed']= delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df.head()
```

```python
# Applying a transformation to the 'player_dismissed' column: 
# If the value is '0', it remains '0', otherwise, it is changed to '1' (indicating the player was dismissed)
delivery_df['player_dismissed']= delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else '1')
```

```python
delivery_df.head()
```

```python
# Converting the 'player_dismissed' column to an integer data type
delivery_df['player_dismissed']= delivery_df['player_dismissed'].astype('int')
```

```python
wicket= delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
wicket
```

```python
# Subtracting the 'wicket' values from 10 and storing the result in a new column 'wicket' in the 'delivery_df' DataFrame
delivery_df['wicket']=10-wicket
```

```python
#Current Run Rate
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
delivery_df.tail()
```

```python
#Required Run Rate
delivery_df['rrr']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
# Function to determine if the batting team won the match
def result(row):
    """
    This function checks if the 'batting_team' is the same as the 'winner'.
    If the batting team won, it returns 1; otherwise, it returns 0.
    Returns:
    int: 1 if the batting team won, 0 otherwise.
    """
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
# Applying the 'result' function to each row of 'delivery_df' 
# and storing the output in a new column 'result'
delivery_df['result']= delivery_df.apply(result, axis=1)
```

```python
delivery_df.head()
```

```python
# Selecting relevant columns from 'delivery_df' and creating a new DataFrame 'final_df'
final_df= delivery_df[['batting_team', 'bowling_team', 'city','runs_left','balls_left', 'wicket','total_runs_x', 'crr', 'rrr','result' ]]
```

```python
final_df
```

```python
# Generating summary statistics for numerical columns in 'final_df'
final_df.describe()
```

```python
# Randomly shuffling all rows of 'final_df' to ensure randomness in the dataset
final_df=final_df.sample(final_df.shape[0])
```

```python
# Selecting a random row from 'final_df' and displaying it
final_df.sample()
```

```python
# Replacing infinite values with NaN in 'crr' to handle division errors
final_df['crr'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Removing any rows with missing (NaN) values from 'final_df'
final_df.dropna(inplace=True)
```

```python
# Filtering out rows where 'balls_left' is zero to avoid division errors or irrelevant data
final_df = final_df[final_df['balls_left'] != 0]
```

```python
# Generating summary statistics for numerical columns in 'final_df'
final_df.describe()
```

```python
x = final_df.iloc[:, :-1]  # Recreate the features dataframe
y = final_df.iloc[:, -1]   # Recreate the target series
```

```python
print(len(x))  # Should print the number of rows in X (features)
print(len(y))  # Should print the number of rows in y (target)
```

```python
# Splitting the dataset into training and testing sets
# 80% of the data will be used for training (x_train, y_train)
# 20% will be used for testing (x_test, y_test)
# random_state=42 ensures reproducibility of the split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
x_train
```

```python
# Display a summary of the x_train DataFrame, including the number of rows, columns, 
# data types, non-null values in each column, and memory usage.
x_train.info()
```

```python
# Import ColumnTransformer to apply different preprocessing steps to specific columns.
# Import OneHotEncoder to convert categorical variables into numerical format using one-hot encoding.
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Define a ColumnTransformer to apply OneHotEncoding to categorical columns.
# - 'batting_team', 'bowling_team', and 'city' are transformed using OneHotEncoder.
# - `sparse=False` ensures the output is a dense NumPy array instead of a sparse matrix.
# - `drop='first'` removes the first category in each column to avoid multicollinearity.
# - `remainder='passthrough'` keeps the remaining columns unchanged.
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
# Import LogisticRegression, a linear model used for classification tasks.
from sklearn.linear_model import LogisticRegression

# Import Pipeline to chain multiple preprocessing steps and a model into a single workflow.
from sklearn.pipeline import Pipeline
```

```python
# Create a machine learning pipeline that automates preprocessing and model training.
# - 'step1': Apply the ColumnTransformer (`trf`) to preprocess categorical features using OneHotEncoding.
# - 'step2': Train a Logistic Regression model with the 'liblinear' solver, 
#   which is efficient for small datasets and binary classification tasks.
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
# Train the pipeline on the training data.
# - The ColumnTransformer ('step1') applies OneHotEncoding to categorical features.
# - The LogisticRegression model ('step2') learns patterns from the transformed data
pipe.fit(x_train, y_train)
```

```python
# Use the trained pipeline to make predictions on the test data.
# - The pipeline first applies the necessary preprocessing (OneHotEncoding).
# - Then, the trained Logistic Regression model predicts the target values.
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred )
```
